```python
# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
# Estimation statistics involves using sample data to estimate population parameters. It provides an inference about the population from which the sample is drawn.

# Point Estimate: A single value estimate of a population parameter (e.g., sample mean is a point estimate of the population mean).
# Interval Estimate: A range of values within which the population parameter is expected to lie, with a certain degree of confidence (e.g., confidence interval).
```

```python
# Q2: Write a Python function to estimate the population mean using a sample mean and standard deviation.
import numpy as np
import scipy.stats as stats

def estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
    return confidence_interval

# Example usage:
sample_mean = 500
sample_std = 50
sample_size = 50
confidence_level = 0.95

confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size, confidence_level)
print(f"95% confidence interval for the population mean: {confidence_interval}")
```

```python
# Q5: Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.
from scipy import stats
import numpy as np

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)
    if p_value < alpha:
        return f"Reject the null hypothesis (p-value: {p_value})"
    else:
        return f"Fail to reject the null hypothesis (p-value: {p_value})"

# Example usage:
np.random.seed(0)
sample1 = np.random.normal(70, 10, 30)
sample2 = np.random.normal(65, 8, 40)

result = two_sample_t_test(sample1, sample2)
print(result)
```

```python
#Q10: Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.
from scipy import stats
import numpy as np

def independent_two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)
    return t_stat, p_value, p_value < alpha

# Example usage:
np.random.seed(0)
sample1 = np.random.normal(70, 10, 30)
sample2 = np.random.normal(65, 10, 30)

t_stat, p_value, reject_null = independent_two_sample_t_test(sample1, sample2)
print(f"t-statistic: {t_stat}, p-value: {p_value}, reject null hypothesis: {reject_null}")
```
